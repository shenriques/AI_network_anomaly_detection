## Testing with service and flags

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
from sklearn.preprocessing import LabelEncoder
```

```python
with_service_flags_cols = ['protocol_type', 'service', 'flag','land', 'urgent', 'num_failed_logins', 'logged_in','root_shell', 'num_shells', 'num_access_files', 'num_outbound_cmds',
 'is_host_login', 'is_guest_login', 'serror_rate', 'srv_serror_rate', 'same_srv_rate', 'dst_host_count', 'dst_host_srv_count', 'dst_host_same_srv_rate',
 'dst_host_serror_rate','class', 'src_bytes', 'dst_host_diff_srv_rate', 'srv_rerror_rate', 'diff_srv_rate', 'srv_count', 
 'dst_host_rerror_rate','dst_host_srv_diff_host_rate', 'dst_host_srv_rerror_rate','count', 'dst_host_srv_serror_rate', 
 'srv_diff_host_rate', 'dst_bytes', 'duration', 'dst_host_same_src_port_rate', 'rerror_rate']

test_data_w_sf = pd.read_csv('data/test_data.csv', usecols=with_service_flags_cols)
test_data_w_sf.head()
```

```python
log_cols = ['src_bytes', 'dst_host_diff_srv_rate', 'srv_rerror_rate', 'diff_srv_rate', 'srv_count', 'dst_host_rerror_rate',
            'dst_host_srv_diff_host_rate', 'dst_host_srv_rerror_rate','count', 'dst_host_srv_serror_rate', 'srv_diff_host_rate',
            'dst_bytes', 'duration', 'dst_host_same_src_port_rate', 'rerror_rate']

for col in log_cols:
    new_col = f"{col}_log" 
    test_data_w_sf[new_col] = np.log(test_data_w_sf[col] + 1)
    
test_data_w_sf = test_data_w_sf.drop(columns=log_cols)
```

```python
test_data_w_sf['class'] = test_data_w_sf['class'].replace({'anomaly': 1, 'normal': 0})
```

```python
le = LabelEncoder()
    
test_data_w_sf['protocol_type'] = le.fit_transform(test_data_w_sf['protocol_type'])
test_data_w_sf['service'] = le.fit_transform(test_data_w_sf['service'])
test_data_w_sf['flag'] = le.fit_transform(test_data_w_sf['flag'])
```
